```python
import random
import matplotlib.pyplot as plt

# Number of steps
n = 15

# Simulate a random walk
def random_walk(n, p):
    positions = [0]

    for _ in range(n):
        step = 1 if random.random() < p else -1
        positions.append(positions[-1] + step)

    return positions


# Generate two walks with different movement probabilities
walk_1 = random_walk(n, p=1/2)
walk_2 = random_walk(n, p=1/4)

times = list(range(n + 1))

# Create the two graphs
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# 1. p = 1/2
axes[0].plot(times, walk_1, linestyle="--", color="black", marker="o", markerfacecolor="hotpink")
axes[0].set_title(r"1-D Random Walk for $n=15$ time steps, $p=1/2$")
axes[0].set_xlabel("Time step")
axes[0].set_ylabel("Position")
axes[0].set_xticks([0, 5, 10, 15])

#2. p = 1/4
axes[1].plot(times, walk_2, linestyle="--", color="black", marker="o", markerfacecolor="hotpink")
axes[1].set_title(r"1-D Random Walk for $n=15$ steps, $p=1/4$")
axes[1].set_xlabel("Time step")
axes[1].set_ylabel("Position")
axes[1].set_xticks([0, 5, 10, 15])

plt.tight_layout()
plt.savefig("random_walk_paths.png", dpi=300, bbox_inches="tight")
plt.show()
```
