```python
# Import initialisers.
import numpy as np
import matplotlib.pyplot as plt
from math import comb

# Define probability function.
def prob(t, j):
    if abs(j) > t or (t + j) % 2 != 0: # If the target node value is larger than the
                                       # number of time steps, or if the target node value
                                       # plus the number of time steps is not even,
                                       # return zero.
        return 0
    n = (t + j) // 2                   # Number of right moving steps to hit target node j.
    return comb(t, n) * (0.5 ** t)     # Return binomial distribution, (0.5) as symmetric.


# Define the number of position and selected time steps
positions = np.arange(-10, 11)
times = [0, 5, 10, 25, 50, 100]

plt.figure(figsize=(8,5))

for t in times:
    probs = [prob(t, j) for j in positions] # Calculate the probability for all positions 
                                            # across the chosen time steps.
    
    plt.plot(positions, probs, label=f"$t={t}$")  # PLot as a graph.

plt.xlabel("Target node $j$")
plt.ylabel(r"Probability $P(X_t=j \mid X_0=0)$")
plt.title("Position vs Probability for a Classical Random Walk")
plt.legend(title="Number of time steps")
plt.grid()
plt.show()
```
